```python
import requests

base_url = "http://localhost:11434/api/chat"

chat_history = [
    {"role": "user", "content": "Hi Ollama!"},
    {"role": "assistant", "content": "Hello! How can I help you today?"}
]

new_message = "Can you tell me a joke?"
messages = chat_history + [{"role": "user", "content": new_message}]

generation_payload = {
    "model": generation_model_id,
    "messages": messages,
    "options": {
        "temperature": 0.1,
        "num_predict": 64},
    "stream": False
}

response = requests.post(base_url, json=generation_payload)
print(response)
if response.status_code == 200:
    data = response.json()
    print("Ollama response:", data.get("message", {}).get("content"))
else:
    print("Error:", response.status_code, response.text)
```

```python
import requests

base_url = "http://localhost:11434/api/embed"


chat_history = [
    {"role": "user", "content": "Hi Ollama!"},
    {"role": "assistant", "content": "Hello! How can I help you today?"}
]

new_message = "Can you tell me a joke?"
messages = chat_history + [{"role": "user", "content": new_message}]

embedding_payload = {
    "model": embedding_model_id,
    "input": "Hi there"
}

response = requests.post(base_url, json=embedding_payload)
if response.status_code == 200:
    data = response.json()
    print("Ollama response:", data.get("embeddings", [None])[0] if data.get("embeddings") else None,
          sep='\n')
else:
    print("Error:", response.status_code, response.text)
```
